## Shifting(Leading and Lagging) Data

```
ts
```

```
ts.shift(2) #ËøôÂíåpremierÁöÑÊªöËΩ¥ÂæàÂÉèÔºåÂæÄÂâçÊå™
```

‰∏äÈù¢ÁöÑshiftÊ≤°ÊúâÊåáÂÆöshiftÂì™ÈáåÔºåÊâÄ‰ª•ÊòØÁßªÂä®Êï∞ÊçÆ„ÄÇÂè™Ë¶ÅÁî®freqÊù•specifyÔºåÂ∞±‰∏ç‰ºöshiftÊï∞ÊçÆÔºåËÄåÊòØshift timestamp

### Shifting data with offsets

ÊúâÊ≤°Êúâ‰∏ÄÁßçÊñπÊ≥ïÊòØËÉΩÂÉèÂä†Âáè‰πòÈô§‰∏ÄÊ†∑ÂØπÊó∂Èó¥ËøõË°åÁßªÂä®Ôºü

Â¶Ç‰∏ã

```
from pandas.tseries.offsets import Day, MonthEnd
```

```
now = datetime(2011, 11, 17)
```

```
now + 3*Day() #ÂæÄnowÊé®‰∏âÂ§©
```

```
now + MonthEnd() #ÂæÄnowÊé®Âà∞‰∏ã‰∏™ÊúàÊú´ÁöÑÂ∑•‰ΩúÊó•
```

```
now + MonthEnd() #ÂæÄnowÊé®Âà∞2‰∏™ÊúàÊú´ÁöÑÂ∑•‰ΩúÊó•
```

ËøòÊúâ‰∏ÄÁßçÂè´ÂÅörollÁöÑÊñπÊ≥ï

```
offset = MonthEnd()
```

```
offset.rollforward(now) #nowÂæÄÂâçrollÂà∞ÊúÄËøëÁöÑÊúàÊú´Â∑•‰ΩúÊó•
```

```
offset.rollback(now) #nowÂæÄÂêérollÂà∞ÊúÄËøëÁöÑÊúàÊú´Â∑•‰ΩúÊó•
```

ÂÖ∂ÂÆûËøôÁßçrollÁöÑÊñπÊ≥ïÂíågroupbyÊê≠ÈÖçÊúÄ‰∏∫üêÇüç∫

```
ts = pd.Series(np.random.randn(20), index=pd.date_range('1/15/2000', periods=20, freq='4d'))
ts
```

```
#Â∞±ÊòØÊääÊØè‰∏™indexÈÉΩÂºÑÊàêÊúàÊú´Â∑•‰ΩúÊó•ÔºåÈÇ£Â∞±‰ºöÂá∫Áé∞‰∏ãÈù¢ÁöÑ‰∏âÁªÑÔºå
#ÁÑ∂ÂêéÊ†πÊçÆËøô‰∏âÁªÑÂèñmean
ts.groupby(offset.rollforward).mean()
```

```
#‰ΩÜÂÖ∂ÂÆûÔºåËøôÁßç‰∏•Ê†ºÊÑè‰πâÊù•ËØ¥ÊòØresampleÔºåÊú¨Êù•Êó∂Èó¥Áª¥Â∫¶ÊòØÂ§©ÔºåÁé∞Âú®Âèñ‰∏∫ÊúàÔºåpandasÊúâÊõ¥Âä†Áõ¥ËßÇÁöÑÂÅöÊ≥ïÔºö
ts.resample('M').mean()
```
