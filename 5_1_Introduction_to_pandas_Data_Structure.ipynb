```
import numpy as np
import pandas as pd
```

## DataFrame

```
data = {'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'],
     'year': [2000, 2001, 2002, 2001, 2002, 2003],
     'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}
frame = pd.DataFrame(data)
frame #随手定义一波dataframe
```

```
frame = pd.DataFrame(data, columns=['year', 'state', 'pop']) #可以自定义column出现的顺序
frame
```

A column in a DataFrame can be retrived as a Series either by **dict-like notation** or by attribute

前者好理解，就是输入df['b']的东西，后者则是df.b  注意！！后者这种only available当是个好命名，比方说df['state a']是不可以写成df.state a的

```
frame['debt'] = np.arange(6)#用这个方法能弄一串数
frame
```

```
del frame['debt']  #删除某column,也可以用 df['xxx'].pop来删除
frame
```

注意，当df.values的时候，values是不同的dtype的话，那么则会统一为object data type
